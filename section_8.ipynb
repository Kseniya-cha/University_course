##### Последовательность Фибоначчи

F_0 = 1, F_1 = 1, F_n = F_(n-1)+F_(n-2)

```python
def f(n):
    # терминальная ветка (явный случай)
    if n == 0 or n == 1:
        return 1
    
    # рекурсивная ветка
    return f(n-1)+f(n-2)

f(3), f(5), f(7)
```

```python
def is_palindrom(s):
    if len(s)<2:
        return True
    
    if s[0] == s[-1]:
        return is_palindrom(s[1:-1])
    else:
        return False

is_palindrom('abacaba'),is_palindrom('abab')
```

##### Поиск элементов (разделяй и властвуй)

```python
def mergesort(lst):
    # 1. терминальная часть
    if len(lst)<2:
        return lst
    elif len(lst) == 2:
        if lst[0]>lst[1]:
            return [lst[1],lst[0]]
        else:
            return [lst[0],lst[1]]
    
    # 2. разбиение задачи
    left = mergesort(lst[:len(lst)//2])
    right = mergesort(lst[len(lst)//2:])

    # 3. объединение подзадач
    li, ri = 0, 0
    result = list()

    while li < len(left) and ri < len(right):
        if left[li] < right[ri]:
            result.append(left[li])
            li += 1
        elif left[li] > right[ri]:
            result.append(right[ri])
            ri += 1
        else:
            result.append(left[li])
            result.append(right[ri])
            li += 1
            ri += 1

    # остатки из списков, если ri/li вышли за предел
    while li < len(left):
        result.append(left[li])
        li += 1
    while ri < len(right):
        result.append(right[ri])
        ri += 1
    
    return result

lst = mergesort([12,21,4,45,2,85,6,102,45])
```

##### Бинарный поиск

Рекурсивный способ

```python
# lst должен быть отсортирован!!!
def binary_search(lst,n):
    if len(lst) == 0:
        return False
    elif len(lst) == 1:
        return lst[0] == n
    
    mid = len(lst)//2
    if lst[mid] == n:
        return True
    if n < lst[mid]:
        return binary_search(lst[:mid], n)
    else:
        return binary_search(lst[mid+1:], n)
    

print(binary_search(lst, 4))
print(binary_search(lst, 1))
```

Итерактивный способ

```python
def binary_search_iter(lst, n):
    l, r = 0, len(lst)-1
    while l <= r:
        mid = l + (r-l)//2
        if lst[mid] == n:
            return True
        elif lst[mid] < n:
            l = mid + 1
        else:
            r = mid - 1
    
    return False


print(binary_search_iter(lst, 4))
print(binary_search_iter(lst, 1))
```

### Динамическое программирование

Подзадачи могут накладываться друг на друга

```python
# ускорить работу для больших чисел
# для последовательности Фибоначчи
# меморизация

cache = {0:1, 1:1}

def f_cahced(n):
    global cache
    if n in cache:
        return cache[n]
    cache[n] = f_cahced(n-1)+f_cahced(n-2)
    return cache[n]

# функция лучше

def dp_f(n):
    dp = [0]*(n+1)
    dp[0], dp[1] = 1, 1
    for i in range(2,n+1):
        dp[i] = dp[i-1] + dp[i-2]
    return dp[n]


print(dp_f(10))
print(dp_f(100))
```

Оптимальное разбиение прута с точки зрения стоимости.

Стоимость не зависит от длины прутиков, на которые разбивается прут.

```python
import random as r
```

```python
n = 100 # длина прута
p = list(range(1,100))*n*2 # цена кусочков
r.shuffle(p)
p = p[:(n+1)] # p[1] - цена кусочка длиной 1 см и тд
p[0] = 0
print(p)
```

Меморизация

```python
cache = {0:0,1:p[1]} # прут длины 1 см разбить нельщя

def maxprice(p,n):
    global cache
    if n == 0:
        return p[0]
    elif n == 1:
        return p[1]
    elif n in cache:
        return cache[n]
    
    # qmax - макс. стоимость разбиения прута
    qmax = p[n] # если прут не делится

    # прут делится на 2 части
    for i in range(1,n):
        pi = p[i] + maxprice(p, n-i)
        if pi > qmax:
            qmax = pi
    cache[n] = qmax
    return qmax

print(p)
maxprice(p, 100)
```

Другой способ (лучше)

```python
def maxrod(p,n):
    dp = [0,p[1]]+[0]*(n-1)
    for i in range(2,n+1):
        # прут не делится
        qmax = p[i]
        # всевозможные деления на две части
        for j in range(1,i):
            if dp[j]+dp[i-j] > qmax:
                qmax = dp[j]+dp[i-j]
        dp[i]=qmax
    return dp[n]

maxrod(p,n)
# 30 минута - найти разбиение
```

Сделка однократная, сначала актив один раз покупается и через какое-то время продается.
Если сделок с положительным финансовым результатом нет, то вывести 0.0

Ввод: В первой строке подается одно число 2 < N ≤ 100000. Далее следует одна строка, в которой задается последовательность цен на актив в течение интересующего Петра периода времени длины N действительных чисел 1.0 ≤ ai ≤ 500.0.

Вывод: Надо вывести финансовый результат самой выгодной сделки с округлением до одного знака после запятой, либо 0.0, если такую сделку невозможно осуществить.

```python
#N = int(input())
#a = input().split()
N = 6
a = '5.0 4.0 3.0 1.0 2.0 9.0'.split()
a = list(map(float,a))


def f(a, N):

    qmin = a.index(min(a))
    lst = [0]

    for i in range(qmin,N-1):
        for j in range(i+1,N):
            if a[j]-a[i] > max(lst):
                lst.append(a[j]-a[i])

    if lst == [0]:
        print(0.0)
    else:
        print(round(max(lst),1))


f(a, N)
```

```python
import sys
 

# Функция для расчета максимальной разницы между двумя элементами в
# Список #, в котором элемент меньшего размера появляется перед элементом большего размера.
def getMaxDiff(A,n):
 
    diff = -sys.maxsize
    max_so_far = A[n - 1]
 
    # обход списка справа и отслеживание максимального элемента
    for i in reversed(range(n - 1)):
 
        # обновляет `max_so_far`, если текущий элемент больше, чем
        # максимальный элемент
        if A[i] >= max_so_far:
            max_so_far = A[i]
 
        #, если текущий элемент меньше максимального элемента,
        #, затем при необходимости обновите разницу.
        else:
            diff = max(diff, max_so_far - A[i])
 
    # обратная разница
    return round(diff,1)
 
#N = int(input())
#a = input().split()
N = 5
a = '10.0 9.0 8.0 7.0 6.0'.split()
a = list(map(float,a))

diff = getMaxDiff(a,N)
if diff < 0:
    print(0.0)

if diff != -sys.maxsize:
    print(diff)
```

Спекулируя на Плюке, Уэф и Би повидали много торговцев гравицапой. Однажды у одного покупателя они увидели необычное устройство. Оно представляло из себя пластинку, на которой было высечено по M целых чисел, в каждой из N строк. Нужна была она ему с такой целью: продавец, продавая гравицапу, должен сам выбрать цену сделки, отметив в каждой строке таблицы ровно по одному числу, но покупатель согласится на сделку только в том случае, если индексы выбранных чисел в строках находятся в порядке неубывания. Например, если есть таблица 3×3 и в первой и второй строке отмечены первые числа строки, а в третьей - второе, то сделка состоится, но если в первой строке выбрано второе число, во второй - первое, то не состоится. Цена сделки равна сумме всех чисел в каждой строке, левее отмеченного числа в строке и плюс сами отмеченные числа.

Итак, Уэф и Би рискуя снова попасть в эцих за жульничество хотят заранее узнать, какую максимальную цену можно установить, используя данную им пластинку.

Вывод: В единственной строке выведите число – максимальную цену сделки, которую можно установить, используя данную пластинку.
