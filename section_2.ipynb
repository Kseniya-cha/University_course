##### Поиск минимума и максимума

```python
n = 1
minimum = None # не установлен
maximum = None
while n < 100:
    n+=int(m.sqrt(n))
    a = int(m.sin(n/(m.pi*m.e))*20)
    # если не установлен или превышает а
    if minimum == None or minimum > a:
        minimum = a
        # обозначим значения а, которые обновляют минимум
        print('*min*', end='')
    if maximum == None or maximum < a:
        maximum = a
        # обозначим значения а, которые обновляют максимума
        print('*max*', end='') 
    print(a, end=' ') # печать в одну строчку

print() # чтоб печаталось с новой строки
print('minimum is', minimum)
print('maximum is', maximum)
```

##### Поиск суммы делителей положительных чисел

```python
n = 1
while n < 100:
    n += int(m.sqrt(n))
    a = int(m.sin(n/(m.pi*m.e))*20)
    if a < 0:
        continue
    d, s = 1, 0 # d - делители, s - их сумма
    while d <= a:
        if a % d == 0:
            s += d
        d += 1
    print('сумма всех делителей числа', a,
    "равна", s) # печать в одну строчку
```

### Тест

Через сколько вызовов функция dice_roll() вернет значение 6 в третий раз?

```python
k = 1

def dice_roll():
    global  k
    n = k
    m = 8
    a = 1664525
    c = 1013904223
    r = 2015
    while n > 0:
        r = (a * r + c) % m
        n = n - 1
    k = k + 1
    return (r % 6) + 1

# решение
counter_all, counter_6 = 0, 0
while True:
    ccc = dice_roll()
    counter_all += 1
    if ccc == 6:
        counter_6 +=1
    if counter_6 ==3:
        break
print(counter_all)
```

Элементы могут принимать целые значения от -10 000 до 10 000 включительно. Напишите программу, позволяющую найти и вывести количество пар элементов, в которых хотя бы одно число делится на 7. В данной задаче под парой подразумевается два подряд идущих элемента.

Не вижу, где проблема.

```python
N = int(input())
first_list = [int(input()) for i in range(N)]
counter = 0

for i in range(0,len(first_list)):
    if first_list[i] % 7 == 0:
        counter += 2
    
        if i+1 < len(first_list) and first_list[i+1] % 7 == 0:
            counter -= 1
        elif i-1 > 0 and first_list[i-1] % 7 == 0:
            counter -= 1
        elif i-1 < 0 or i+1 > len(first_list)-1:
            counter -= 1

print(first_list)
print(counter)
```

### C. Сумма цифр, кратных 4

На обработку поступает натуральное число, не превышающее 10^9.Нужно написать программу, которая выводит на экран сумму цифр числа, которые делятся на четыре. Если в числе нет таких цифр, требуется на экран вывести No.

```python
N = input()

summa = 0
summa_zero = 0
for num in N:
    num = int(num)
    if num == 0:
        summa_zero += 1
    elif num % 4 == 0:
        summa += num

if summa == 0 and summa_zero == 0:
    print('No')
else:
    print(summa)
```

### D. Составные целые числа

На вход подается целое число A, не превосходящее 10^9. Каждое такое число можно разложить на множители A=B∗C, вероятно, несколькими способами. Найти и вывести минимально возможное значение суммы B+C.

Например, если A=12345, то

12345=1∗12345, следовательно B+C=1+12345=12346

12345=3∗4115, следовательно B+C=3+4115=4118



Работает дольше 1 секунды.

```python
def find_min(A):
    minimum = None
    for b in range(1,A//2):
        for c in range(1,A+1):
            if c*b == A:
                if minimum == None or minimum > c+b:
                    minimum = c+b
    print(minimum)

#A = int(input())
find_min(11111)
```

### E. Ближайшие простые числа

На вход подается натуральное число A, не превосходящее 10000. Вам надо найти ближайшие к нему простые числа B и C, такие что B<=A<=C.

```python
def test_prime(n):
    if (n==1):
        return False
    elif (n==2):
        return True;
    else:
        for x in range(2,n):
            if(n % x==0):
                return False
        return True     

A = int(input())
B, C = A, A

if not test_prime(A):
    while True:
        B -= 1
        if test_prime(B):
            break
    while True:
        C += 1
        if test_prime(C):
            break

print(B, C)
```
