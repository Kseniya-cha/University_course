Найти число в списке B, ближайшее к среднему арифметическому значению его элементов.

```python
B = [1,2,4,3,8,6,7,1,5,3,4,2,7]

mid = round(sum(B)/len(B)) # среднее арифметическое

b_minus = []
for i in range(len(B)):
    # модуль разности элементов и среднего арифметического
    b_minus.append([abs(mid-B[i]),B[i]])
    if abs(mid-B[i]) >= abs(mid-B[i-1]):
        continue
    # сокращаем тайминг
    if abs(mid-B[i]) == 0:
        break

b = min(b_minus)[1]
print(b)
```

Простые числа.

```python
from math import ceil, sqrt

def prime(N):
    if N % 2 == 0 or N < 2:
        return False
    # перебираем нечётные числа с шагом 2 до корня N
    for x in range(3,ceil(sqrt(N))+1,2):
        if N % x == 0:
            return False
    return True

prime(7), prime(6), prime(2), prime(1)
```

Задача: найти а: a_1^2 + a_2^2 + ... + a_m^2 = a_(m+1)^2

```python
m = 2
A = [a**2 for a in range(1,100)]
# product - возвращает итератор
for x in it.product(A,repeat=m):
    # repeat указывает: длина кортежей = m, элементы из А
    s = sum(x)
    a = math.floor(math.sqrt(s)) # целая часть от sqrt от s
    if s == a**2:
        print([math.floor(math.sqrt(xi)) for xi in x], a)
```

Поиск контрпримера к гипотезе Эйлера: n-ая степень какого-либо числа должна состоять из n слагаемых n-ой степени. Покажем, что хватит n-1 или меньше слагаемых

```python
n = 3
while True:
    A = [a**n for a in range(200)]
    print('n =', n)
    for x in it.product(A, repeat=n-1):
        if x.count(0) > len(x)-2:
            continue
        s = sum(x)
        a = math.floor(math.pow(s, 1/n))
        if s == a**n:
            print([math.floor(math.pow(xi, 1/n)) for xi in x], a)
            exit()
    n += 1
```

##### Перебор сочетаний

Есть набор предметов, которые можно поместить в рюкзак. У предметов есть масса и цена. Рюкзак может вместить M единиц массы.

Нужно поместить в рюкзак 3 предмета.

```python
A = ['крыло', "гриб", "русалка", "перо", "ведро", "лягушка"]
weight = [7,3,10,4,7,3]
price = [14,10,17,12,5,10]

M = 20
maxvalue = -1
maxprice = None
maxnames = None

for i in range(len(A)):
    for j in range(i+1,len(A)):
        for k in range(j+1,len(A)):
            # s - сколько сейчас в рюкзаке, w - сколько ещё влезет
            s, w = 0, M
            for m in [i,j,k]:
                w -= weight[m]
                s += price[m]
            if w >= 0 and s > maxvalue:
                maxprice = (i,j,k)
                maxnames = (A[i], A[j], A[k])
                maxvalue = s
print(maxvalue,maxprice,maxnames)
```

Найти максимальную стоимость вмещаемых предметов. Число предметов - любое.

```python
maxk = min(len(A), M//min(weight)+1)
for k in range(1,maxk+1): # k - количество вмещаемых предметов
    # combinations - для перебора сочетаний, передаётся
    # множество (индексы А), k - длина наборов
    for items in it.combinations(range(len(A)), k):
        s = sum([price[i] for i in items])
        w = M-sum([weight[i] for i in items])
        if w >= 0 and s > maxvalue:
            maxprice = items
            maxnames = [A[i] for i in items]
            maxvalue = s
print(maxvalue,maxprice,maxnames)
```

Если предметы могут повторяться:

```python
maxk = min(len(A), M//min(weight)+1)
for k in range(1,maxk+1):
    # combinations_with_replacement - с повторением
    for items in it.combinations_with_replacement(range(len(A)), k):
        s = sum([price[i] for i in items])
        w = M-sum([weight[i] for i in items])
        if w >= 0 and s > maxvalue:
            maxprice = items
            maxnames = [A[i] for i in items]
            maxvalue = s
print(maxvalue,maxprice,maxnames)
```

### D. Рюкзак из трёх предметов

Перед вами находится несколько предметов, обладающих стоимостью и весом. Вам нужно определить максимальную стоимость набора из трёх предметов, которые можно положить в рюкзак, способный выдержать выдержать предметы общим весом M.

Ввод: на вход подается три строки. В первой строке набор стоимостей предметов, во второй набор их весов. Количество предметов больше или равно трем и не превосходит 100. Третья строка содержит максимальный вес M, который способен выдержать рюкзак.

Вывод: Число, показывающее максимальную стоимость рюкзака состоящего из трёх предметов. Если ни одного набора из трёх предметов не вмещается в рюкзак, вывести -1.

```python
pricein = ('3 4 1 5').split()
weightin = ('2 5 6 3').split()
M = 11

price = [int(pricein[i]) for i in range(len(pricein))]
weight = [int(weightin[i]) for i in range(len(weightin))]

maxvalue = -1

for i in range(len(weight)):
    for j in range(i+1,len(weight)):
        for k in range(j+1,len(weight)):
            s, w = 0, M
            for m in [i,j,k]:
                w -= weight[m]
                s += price[m]
            if w >= 0 and s > maxvalue:
                maxvalue = s
print(maxvalue)
```

```python
pricein = ('3 4 1 5').split()
weightin = ('2 5 6 3').split()
M = 11

price = [int(pricein[i]) for i in range(len(pricein))]
weight = [int(weightin[i]) for i in range(len(weightin))]

maxvalue = -1

for items in it.combinations(range(len(weight)), 3):
    s = sum([price[i] for i in items])
    if (M-sum([weight[i] for i in items])) >= 0 and s > maxvalue:
        maxvalue = s

print(maxvalue)
```
