### A. Файловый шифратор

Ваша цель состоит в шифровании нескольких сообщений.

Для решения этой задачи вам предлагается ключ, который ставит в соответствие буквам латинского алфавита и их шифрованный аналог. Например, a -> b, b -> a, c -> z, и так далее. Вам надо заменить буквы в исходном сообщении согласно этому ключу. Аналогичная замена производится в случае, если будут встречаться заглавные буквы. Остальные символы сообщений не шифруются.

```python
file1 = open('input1.txt', 'r')

cipher = file1.readline(26)
text = file1.read()

alphabet = 'abcdefghijklmnopqrstuvwxyz'
text_new = ''
d = dict()
for i in range(26):
    d[alphabet[i]] = cipher[i]

for elem in text:
    if elem.isalpha():
        if elem.isupper():
            text_new += (d[elem.lower()]).upper()
        else:
            text_new += d[elem.lower()]
    else:
        text_new += elem
file1.close()

text_new = text_new.strip()

file2 = open('output1.txt', 'w')
file2.write(text_new)
file2.close()
```

### B. Кофеманы

В кофейни города N приходят посетители и заказывают кофе. Каждый посетитель может приходить несколько раз и делать несколько заказов.

Найдите посетителей, которые ходят во все кофейни.

```python
file1 = open('input2.txt', 'r')

coffee = list()
clients = list()
file_lst = list()
d = dict()
client_str = ''

while True:
    line = file1.readline().strip()
    clients.append(line[3:])
    coffee.append(line[:2])
    if not line:
        break
file1.close()

clients.remove('')
coffee.remove('')

count_coffee = len(set(coffee))

for i in range(len(clients)):
    try:
        if (d[clients[i]] is not None) and (coffee[i] not in d[clients[i]]):
            d[clients[i]] += ' '+ coffee[i]
    except:
        d[clients[i]] = coffee[i]

count = 0
for k in d.keys():
    if len(d[k].split()) == count_coffee:
        client_str += str(k) + ' '
        count += 1

if count == 0:
    client_str = 'No'

lst = sorted(client_str.split())
client_str = ' '.join(lst)

file1 = open('output2.txt','w')
file1.write(client_str)
file1.close()
```
