Сравнение площадей многоугольников

```python
class Polygon:
    def __init__(self,s,n):
        # сохраняем внутри объекта значения данных переменных
        self._side = s
        self._n = n
    
    def area(self):
        return (self._n*self._side**2)/(4.0*m.tan(m.pi/self._n))

    def __str__(self): # спец метод
        return str(self._n)+'-угольник со стороной '+str(self._side)
    
    def getside(self):
        print('возврат длины стороны')
        return self._side

    def setside(self,value):
        print('попытка установить значение side =', value)
        if value>0:
            self._side = value

    def delside(self):
        print('чё? не трожь стороны!')

    side = property(getside, setside, delside, 'свойство side')

    # метод для сравнения (только для строгих неравенств)
    def __lt__(self, other):
        return self.area() < other.area()
    # нестрогое неравенство
    def __le__(self, other):
        sa = self.area()
        oa = other.area()
        return sa < oa or abs(sa - oa) < 0.001
    # равенство (одинаковое кол-во сторон,
    # их площади отличаются меньше чем на 0.001)
    def __eq__(self, other):
        return self._n==other._n and abs(self._side-other._side)<0.001
    
    # умножение стороны на число
    def __mul__(self, other):
        if other>0:
            # создаём новый объект с изименёнными характеристиками
            return Polygon(self._side*other, self._n)
        else:
            return self


p1 = Polygon(3.5, 10)
p2 = Polygon(14.5, 3)
p3 = Polygon(9.5415, 4)
p4 = Polygon(9.541, 4)
print(p1.area(),p2.area(),p3.area(),p4.area())

print(p1 < p2, p1 > p2, p1 <= p2)
print(p2>p3,p2<p3,p2>=p3,p2<=p3)
print(p2==p3, p3==p4)

# определение многоугольника, стороны которого в 5 раз больше
p_m = p1*5
print(p1, p_m)
p_m = p1*(-10)
print(p1, p_m)
```

#### Итераторы

```python
lst = [1,2,3,10,15,'a']

# обычно
for elem in lst:
    print(elem,type(elem), end='; ')
print()

# через итераторы
# iter - возвращает итератор

lst_it = iter(lst)
print(lst_it, type(lst_it))
while True:
    try:
        # next продвигает итератор и возвращает настоящее значение
        elem = next(lst_it)
        print(elem, type(elem), end='; ')
    # закончились итерации
    except StopIteration:
        break
print()
```

Написание собственного итератора

Пробегает все простые числа в определённом интервале

```python
class PrimeNumbers:
    def __init__(self,a,b=None):
        if b is None:
            self.minn = 1
            self.maxn = a
        else:
            self.minn = a
            self.maxn = b
    
    def nextprime(self,a):
        if a <= 2:
            return 2
        for n in range(a,self.maxn):
            if n%2 == 0:
                continue
            for j in range(3,n//2,2):
                if n%j==0:
                    break
                else:
                    return n
        return self.maxn

    def __iter__(self):
        # current хранит простое число, которое используется
        # на текущей итерации
        self.current = self.minn
        return self

    def __next__(self):
        self.current = self.nextprime(self.current+1)
        if self.current >= self.maxn:
            # принудттельный вызов исключения
            raise StopIteration
        return self.current

for num in PrimeNumbers(10,350):
    print(num, end=' ')
```
