## Лекции

Поиск подстроки

```python
s = 'Who likes to known Python? It is me who  likes to known Python!'

def searching(s, sub):
    print(s, sub in s)
    print('начинает ли строку?',s.startswith(sub))
    # startswith - если sub начинает строку s
    print('заканчивает ли строку?',s.endswith(sub))
    # endswith - если sub заканчивает строку s
    # поиск слева направо, содержится ли sub в s
    si = s.find(sub) 
    print('индекс si =', si)
    # поиск спарва налево
    ri = s.rfind(sub)
    print('индекс ri =', ri)
    print('количество непересекающихся вхождений count =',
    s.count(sub), end='\n\n')

searching(s, 'who')
searching(s, 'Python')
searching(s, 'Pascal')
```

### D. Самый длинный общий префикс

Дан набор из N строк. Вам надо найти самый длинный общий префикс среди этих строк.

##### Мой код

```python
#n = int(input())
#lst = [(input()).lower() for i in range(n)]
n = 3
lst = ['shrimp', 'shrink', 'shrine']
#lst = ['flower', 'flow', 'flight']

len_min = min([len(elem) for elem in lst])
ll = [lst[0][0]]

for i in range(len_min):

    for elem in lst:
        if elem[i] == lst[0][i]:
            a = 1
        else:
            a = 0
            break

    if a == 1:
        ll[0] += lst[0][i]
ll[0] = ll[0][1:]

print(ll[0])
```
